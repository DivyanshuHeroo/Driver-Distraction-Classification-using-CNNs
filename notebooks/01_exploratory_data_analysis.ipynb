# Exploratory Data Analysis — State Farm Distracted Driver

Quick look at class balance and sample images before training.
Run from the project root after the dataset is in `data/imgs/train`.

```python
import os, sys
sys.path.append('..')
import matplotlib.pyplot as plt
from PIL import Image
from src.utils import load_config, get_class_names

config = load_config('../config.yaml')
train_dir = os.path.join('..', config['paths']['data_dir'])
classes = get_class_names(config)
classes
```

## Class distribution

```python
counts = {c: len(os.listdir(os.path.join(train_dir, c))) for c in classes}
plt.figure(figsize=(10, 4))
plt.bar([config['classes'][c] for c in classes], list(counts.values()))
plt.xticks(rotation=45, ha='right')
plt.title('Images per class')
plt.tight_layout(); plt.show()
counts
```

## Sample image per class

```python
fig, axes = plt.subplots(2, 5, figsize=(16, 7))
for ax, c in zip(axes.ravel(), classes):
    folder = os.path.join(train_dir, c)
    img = Image.open(os.path.join(folder, os.listdir(folder)[0]))
    ax.imshow(img); ax.axis('off')
    ax.set_title(f"{c}: {config['classes'][c]}", fontsize=9)
plt.tight_layout(); plt.show()
```
